```python
import pandas as pd
from sklearn.preprocessing import RobustScaler
from sklearn.decomposition import PCA
import pickle
```

```python
df=pd.read_csv("CC GENERAL.csv")
```

```python
df=df.dropna()
```

```python
df.drop('CUST_ID',axis=1,inplace=True)
```

```python
df=df[['BALANCE','PURCHASES','INSTALLMENTS_PURCHASES','CASH_ADVANCE','PAYMENTS','CREDIT_LIMIT']]
```

# Load saved scalar

```python
loaded_scalar=pickle.load(open("Final_scalar.sav",'rb'))
```

# Loading saved model

```python
loaded_model=pickle.load(open("Finalized_ClusterModel.sav",'rb'))
```

```python
#Getting user input
# Dictionary to store user input
user_input = {}
print("Please enter values for the following features:")
for feature in df.columns:
    
        while True:
            try:
                val = float(input(f"{feature}: "))
                user_input[feature] = val
                break
            except ValueError:
                print("Invalid input. Please enter a numeric value.")

 # Convert to DataFrame (single row)
new_customer = pd.DataFrame([user_input])  
```

```python
new_customer
```

# Standardizing user input

```python
new_scaled=loaded_scalar.transform(new_customer)
```

# Making Predictions

```python
cluster_num=loaded_model.predict(new_scaled)[0]
# Map to business-friendly cluster names
cluster_names = {
    0: 'Low spenders',
    1: 'Premium High-Value Customers',
    2: 'Cash Advanceâ€“Dependent Customers',
    
}
cluster_name = cluster_names[cluster_num]

print("\nPredicted cluster number:", cluster_num)
print("Predicted cluster name:", cluster_name)
```
